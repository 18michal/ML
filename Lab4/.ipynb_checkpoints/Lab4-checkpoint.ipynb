### Zadanie 1

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../Lab2/phpMYEkMl.csv', index_col='id')
df = df.replace('?', np.nan)

new_column_name = []
for col in df.columns:
    new_column_name.append(col[1:-1])
df.columns = new_column_name

df['CabinReduced'] = df['cabin'].astype(str).str[0]

df.head(20)
```

Stwórz zmienną do której zapiszesz listę z nazwami trzech kolumn –

```python
col_name = ['CabinReduced', 'cabin', 'sex']
# df[['CabinReduced', 'cabin', 'sex']]
```

### Zadanie 4
Jako zmienną niezależną ustaw dane składające się z kolumn o nazwach
zapisanych w col_name. Jako zmienną zależną ustaw kolumnę mówiącą o
tym czy ktoś przeżył czy nie. Ustaw rozmiar zbioru testowego na 20 lub
30% całości. Parametr random_state ustaw jako 0.
Wyświetl wymiary zbiorów X_train, X_test, y_train, y_test używając
.shape() i skomentuj wyniki.

```python
from sklearn.model_selection import train_test_split

zmienna_niezalezna = df[col_name]
zmienna_zalezna = df['survived']

X_train, X_test, y_train, y_test = train_test_split(zmienna_niezalezna, zmienna_zalezna, test_size=0.2, random_state=0)
print(f'Wymiar X_train: {X_train.shape}')
print(f'Wymiar X_test: {X_test.shape}')
print(f'Wymiar y_train: {y_train.shape}')
print(f'Wymiar y_test: {y_test.shape}')
```

Komentarz wyników
- parametr random_state ustawiony na 0, zapewni powtarzalność wyników podczas kolejnych uruchomień programu,
- zbiór treningowy zawiera 1047 wierszy oraz 3 kolumny
- zbiór testowy zawiera 262 wierszy oraz 3 kolumny
- zbiór treningowy zmiennej zależnej zawiera 1047 wartości
- zbiór testowy zmiennej zależnej zawiera 262 wartości

### Zadanie 5
Zbadaj w zależności od kardynalności danej zmiennej czy rozkład etykiet
dla poszczególnych cech w zbiorach testowych i treningowych jest
równomierny. 

```python
for col in col_name:
    temp = X_train[col].unique()
    unique_test = [x for x in temp if x not in X_test[col].unique()]
    print(f'Kolumna {col}:')
    print(f'Liczba unikalnych wartości w zbiorze treningowym: {len(temp)}')
    print(f'Liczba unikalnych wartości w zbiorze treningowym, które nie występują w zbiorze testowym: {len(unique_test)}', end='\n\n')
```

### Zadanie 6
Wykonaj kodowanie zmiennych kategorycznych do zmiennych
liczbowych. Wykorzystaj pętlę for i metodę enumerate()

```python
def categorical_coding(df, col):
    categorical_data = df[col].unique()
    label_dict = {}
    for i, category in enumerate(categorical_data):
        label_dict[category] = i
    return label_dict

CabinReduced_map_train = categorical_coding(X_train, col_name[0])
cabin_map_train = categorical_coding(X_train, col_name[1])
sex_map_train = categorical_coding(X_train, col_name[2])


CabinReduced_map_test = categorical_coding(X_test, col_name[0])
cabin_map_test = categorical_coding(X_test, col_name[1])
sex_map_test = categorical_coding(X_test, col_name[2])

print('CabinReduced zbiór X_train:')
print(CabinReduced_map_train)
print('\ncabin zbiór X_train:')
print(cabin_map_train)
print('\nsex zbiór X_train:')
print(sex_map_train)
print('\nCabinReduced zbiór X_test:')
print(CabinReduced_map_test)
print('\ncabin zbiór X_test:')
print(cabin_map_test)
print('\nsex zbiór X_test:')
print(sex_map_test)
```

```python
X_train['CabinReduced_map'] = X_train['CabinReduced'].map(CabinReduced_map_train)
X_train['cabin_map'] = X_train['cabin'].map(cabin_map_train)
X_train['sex_map'] = X_train['sex'].map(sex_map_train)
X_test['CabinReduced_map'] = X_test['CabinReduced'].map(CabinReduced_map_test)
X_test['cabin_map'] = X_test['cabin'].map(cabin_map_test)
X_test['sex_map'] = X_test['sex'].map(sex_map_test)

print('Zbior treningowy')
print(X_train.head(20))

print('\nZbior testowy')
print(X_test.head(20))
```

### Zadanie 8
Sprawdź liczbę brakujących wartości w zmodyfikowanych zbiorach.
Zapisz wyniki i skomentuj. 

```python
test_nan_CabinReduced_map = X_test['CabinReduced_map'].isna().sum()
test_nan_cabin_map = X_test['cabin_map'].isna().sum()
test_nan_sex_map = X_test['sex_map'].isna().sum()

train_nan_CabinReduced_map = X_train['CabinReduced_map'].isna().sum()
train_nan_cabin_map = X_train['cabin_map'].isna().sum()
train_nan_sex_map = X_train['sex_map'].isna().sum()

print(f'W zbiorze testowym kolumna CabinReduced_map ma {test_nan_CabinReduced_map} brakujacych wartosci.')
print(f'W zbiorze testowym kolumna cabin_map ma {test_nan_cabin_map} brakujacych wartosci.')
print(f'W zbiorze testowym kolumna sex_map ma {test_nan_sex_map} brakujacych wartosci.')

test_nan = X_test[col_name].isna().sum()
print('\nZbiór testowy dla pozostałych klumn - suma brakujących wartosci w kolumnie')
print(test_nan, end='\n\n')

print(f'W zbiorze treningowym kolumna CabinReduced_map ma {train_nan_CabinReduced_map} brakujacych wartosci.')
print(f'W zbiorze treningowym kolumna cabin_map ma {train_nan_cabin_map} brakujacych wartosci.')
print(f'W zbiorze treningowym kolumna sex_map ma {train_nan_sex_map} brakujacych wartosci.')

train_nan = X_train[col_name].isna().sum()
print('\nZbiór treningowy dla pozostałych klumn - suma brakujących wartosci w kolumnie')
print(train_nan, end='\n\n')
```

W zbiorze testowym oraz treningowym nie ma brakujących wartości w kolumnach, do których zostały dokonane kodowania zmiennych kategorycznych. <br>W zbiorze testowym w kolumnie cabin są 211 brakujących wartości.<br>W zbiorze treningowym w kolumnie cabin są 803 brakujących wartości.<br><br>To sugeruje, że ta kolumna może być trudna do zastosowania w modelowaniu predykcyjnym.

### Zadanie 10
Porównaj ile unikalnych wartości jest w zbiorze treningowym, a ile w
zbiorze testowym (funkcja len).
Jaka jest różnica pomiędzy liczbą etykiet przed i po redukcji oraz
mapowaniu?
Czy cały proces, który został do tej pory wykonany może mieć wpływ na
końcowy wynik predykcji i jakość modelu?

```python
for col in X_train.columns:
    print(f'Kolumna {col}:')
    print('Liczba unikalnych wartości w zbiorze treningowym:', len(X_train[col].unique()))
    print('Liczba unikalnych wartości w zbiorze testowym:', len(X_test[col].unique()), end='\n\n')
```

Różnica pomiędzy liczbą etykiet przed i po redukcji oraz mapowaniu zachodzi tylko w kolumnie cabin,<br>Przed:<br>
  Liczba unikalnych wartości w zbiorze testowym: 30<br>
Po:<br>
  Liczba unikalnych wartości w zbiorze testowym: 49<br><br><br>
Wpływ na jakość predykcji może mieć fakt, że część klas lub wartości została po prostu usunięta lub zmapowana na inną wartość.<br>
Redukcja liczby etykiet może pomóc w uproszczeniu modelu i zwiększeniu jego wydajności. Ponieważ liczba unikalnych wartości zwiększyła się.
